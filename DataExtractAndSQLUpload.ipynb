```python
import os
import json
import pandas as pd
```

```python
##aggregated_transaction

path_1 ="C:/Users/91936/OneDrive/Desktop/Guvi/Capstone Project 2 -Phonepe/pulse/data/aggregated/transaction/country/india/state/"
agg_tran_list = os.listdir(path_1)

columns1 = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "TransactionType":[],
           "TransactionCount":[],
           "TransactionAmount":[]}


for state in agg_tran_list:
    current_state = path_1 + state + "/"
    agg_year_list = os.listdir(current_state)

    for year in agg_year_list:
        current_year = current_state + year + "/"
        agg_file_list = os.listdir(current_year)

        for file in agg_file_list:
            current_file = current_year + file
            data = open(current_file,"r")
            A= json.load(data)


            for i in A["data"]["transactionData"]: 
   
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))

                name = i["name"]
                columns1["TransactionType"].append(name)

                count = i["paymentInstruments"][0]["count"]
                columns1["TransactionCount"].append(count)

                amount = i["paymentInstruments"][0]["amount"]
                columns1["TransactionAmount"].append(amount)

Aggregated_Transaction = pd.DataFrame(columns1)


Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.replace("-"," ")
Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.title()
Aggregated_Transaction['States'] = Aggregated_Transaction['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
  

```

```python
## aggregated_user

path_2 ="C:/Users/91936/OneDrive/Desktop/Guvi/Capstone Project 2 -Phonepe/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path_2)

columns2 = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Brands":[],
           "TransactionCount":[],
           "Percentage":[]}


for state in agg_user_list:
    current_state = path_2 + state + "/"
    agg_year_list = os.listdir(current_state)

    for year in agg_year_list:
        current_year = current_state + year + "/"
        agg_file_list = os.listdir(current_year)

        for file in agg_file_list:
            current_file = current_year + file
            data = open(current_file,"r")
            B= json.load(data)
            try:
                for i in B["data"]["usersByDevice"]: 
    
                    columns2["States"].append(state)
                    columns2["Years"].append(year)
                    columns2["Quarter"].append(int(file.strip(".json")))

                    brand = i["brand"]
                    columns2["Brands"].append(brand)

                    count = i["count"]
                    columns2["TransactionCount"].append(count)

                    percentage = i["percentage"]
                    columns2["Percentage"].append(percentage)
            
            except : 
                pass


Aggregated_Users = pd.DataFrame(columns2)

Aggregated_Users["States"] = Aggregated_Users["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Aggregated_Users["States"] = Aggregated_Users["States"].str.replace("-"," ")
Aggregated_Users["States"] = Aggregated_Users["States"].str.title()
Aggregated_Users['States'] = Aggregated_Users['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
  
```

```python
## Map_User

path_4 ="C:/Users/91936/OneDrive/Desktop/Guvi/Capstone Project 2 -Phonepe/pulse/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path_4)

columns4 = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Districts":[],
           "RegisteredUsers":[],
           "AppOpens":[]}


for state in map_user_list:
    current_state = path_4 + state + "/"
    agg_year_list = os.listdir(current_state)

    for year in agg_year_list:
        current_year = current_state + year + "/"
        agg_file_list = os.listdir(current_year)

        for file in agg_file_list:
            current_file = current_year + file
            data = open(current_file,"r")
            D= json.load(data)


            for i in D["data"]["hoverData"].items(): 
    
                    columns4["States"].append(state)
                    columns4["Years"].append(year)
                    columns4["Quarter"].append(int(file.strip(".json")))

                    name = i[0]
                    columns4["Districts"].append(name)

                    registeredUsers = i[1]["registeredUsers"]
                    columns4["RegisteredUsers"].append(registeredUsers)

                    appOpens = i[1]["appOpens"]
                    columns4["AppOpens"].append(appOpens)
            
Map_User = pd.DataFrame(columns4)

Map_User["States"] = Map_User["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_User["States"] = Map_User["States"].str.replace("-"," ")
Map_User["States"] = Map_User["States"].str.title()
Map_User['States'] = Map_User['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
  
```
